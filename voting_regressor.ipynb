```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("cgpa.csv")
df
```

```python
# x = df[["cgpa"]]
x = df.iloc[:,:-1]
x
```

```python
y = df.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
lr.score(x_train,y_train)*100,lr.score(x_test,y_test)*100
```

```python
lr.predict(x_test)
y_test
```

```python
d1 = DecisionTreeRegressor()
d1.fit(x_train,y_train)
d1.score(x_train,y_train)*100,d1.score(x_test,y_test)*100
```

```python
cl1 = SVR()
cl1.fit(x_train,y_train)
cl1.score(x_train,y_train)*100,cl1.score(x_test,y_test)*100
```

## Ensamle technique

```python
from sklearn.ensemble import VotingRegressor
```

```python
l1= [('ln',LinearRegression()),('dt1',DecisionTreeRegressor()),('svr1',SVR())]
```

```python
v1 = VotingRegressor(l1)
v1.fit(x_train,y_train)
v1.score(x_train,y_train)*100, v1.score(x_test,y_test)*100
```

```python
pred = {'lr':lr.predict(x_test),'dt':d1.predict(x_test),'svm':cl1.predict(x_test),'vr':v1.predict(x_test)}
```
